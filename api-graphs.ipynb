```python
import dotenv
import os
import requests
import pandas as pd

# Load the .env file from the current directory
dotenv.load_dotenv("aquila_graph.env")

# Now you can access variables such as FRED_API_KEY via os.environ
fred_api_key = os.environ.get("FRED_API_KEY")



# Define the FRED API endpoint and parameters
series_id = "AUST448BPPRIV"
url = "https://api.stlouisfed.org/fred/series/observations"
params = {
    "series_id": series_id,
    "api_key": fred_api_key,
    "file_type": "json"
}

# Make the GET request
response = requests.get(url, params=params)
fred_data = response.json()

# Convert the observations to a DataFrame
observations = fred_data.get("observations", [])
df = pd.DataFrame(observations)

# Display the first few rows of the dataframe
df.head()
```
